```python
import os
import pandas as pd
from collections import Counter
import pickle
# 이게 가장 효과과 좋다.
from konlpy.tag import Okt
okt = Okt()
```

```python
output_folder = "text_split"
os.makedirs(output_folder, exist_ok=True)  # 폴더가 없으면 생성

with open('data.txt', 'r', encoding='utf-8') as file:
    for i in range(34):
        content = file.read(10000)  # 매번 10,000자씩 읽기
        if not content:  # 더 이상 읽을 내용이 없으면 종료
            break

        # 파일을 text_split 폴더 안에 저장
        filename = os.path.join(output_folder, f"{i+1}.txt")
        with open(filename, 'w', encoding='utf-8') as new_file:
            new_file.write(content)

        print(f"Created: {filename}")  # 생성된 파일 경로 출력
#336020
# 만자 정도는 인식을 한다.
# -> 파일을 34개로 나누자
# 일단 나누기는 완료
```

```python
# 일단 문장수는 했으니 
# 요소들을 추출 해야 한다.


output_text_processfolder = "text_process"
os.makedirs(output_text_processfolder, exist_ok=True)  # 폴더가 없으면 생성

for i in range(34):
    file = open(f'text_split/{i+1}.txt','r')
    content = file.read()
    sentence = okt.morphs(content)
    print(f"{i+1}번쨰 읽기는 함")
    with open(f"text_process/{i+1}.pkl", "wb") as file:
        pickle.dump(sentence, file)
```

```python
# 이제는 뭐를 해야 할까?
# csv 파일로 저장을 하자

output_csv = "word_counting.csv"


word_counts = Counter()

for sen in range(34):
    with open(f"text_process/{i+1}.pkl", "rb") as file:
        loaded_data = pickle.load(file)
    content = loaded_data
    word_counts.update(content)
# 단어 빈도수를 DataFrame으로 변환
df = pd.DataFrame(word_counts.items(), columns=["단어", "빈도수"])

# 빈도수 기준으로 정렬 (내림차순)
df = df.sort_values(by="빈도수", ascending=False)

# CSV 파일로 저장
df.to_csv(output_csv, index=False, encoding='utf-8-sig')

print(f"✅ 단어 빈도수 분석 완료! CSV 파일 저장: {output_csv}")
```

```python
# 이제 노가다 시작
# 미친거 아니야?
# -> gpt시켜


stopword_list = [
    # 🚨 불필요한 조사 및 접속사
    "은", "는", "이", "가", "을", "를", "의", "에", "로", "한", "있다",
    "것", "들", "수", "과", "으로", "도", "에서", "하는", "가", "이다",
    "한", "다", "같다", "없다", "아닌",
    
    # 🚨 특수문자
    ".", ",", '"', "'", "(", ")", "「", "」", "/", "?", "!", ":", "-",
    
    # 🚨 의미 없는 단어
    "그것", "자기", "때문", "그러나", "어떤", "대해", "위해",
    "한때", "결국", "일부", "처럼", "또한", "않다", "하지"

    # 내가 따로 추가
    "하고","있는","@munhakdongne","트위터","|","mhdn","ccm","naver","cafe","://","955-8864"
    ,"955-8855","editor@munhak.com","으로도","으로는","로는","었다","있다고"
    ,"이었","이었을","리지","하기","955-3576","*","있었다","에게","했다","때",
]

# CSV 파일 불러오기
df = pd.read_csv("stopword.csv")

# 불용어 제거
df_filtered = df[~df["단어"].isin(stopword_list)]  # 불용어가 포함된 단어 제거

# 결과 저장
df_filtered.to_csv("final.csv", index=False)

print("✅ 불용어 제거 완료! 결과 저장: final.csv")
```
