```python
# 파일 읽기
file = open('data.txt','r')
content =file.read()

# Okt 객체 생성
okt = Okt()

# 문장 단위로 나누기
sentences = re.split(r'[.!?]', content)

# 각 문장을 형태소 분석하여 리스트에 저장
morphed_sentences = [okt.morphs(sentence.strip()) for sentence in sentences if sentence.strip()]

# 결과 파일 저장
with open('everything.txt', 'w') as output_file:
    for i, morphs in enumerate(morphed_sentences, 1):
        output_file.write(f"{str(morphs)}\n")

print("저장 완료")
```

```python
# 파일 읽기
file = open('data.txt','r')
content =file.read()

# Okt 객체 생성
okt = Okt()

# 문장 단위로 나누기
sentences = re.split(r'[.!?]', content)

# 각 문장을 형태소 분석하여 리스트에 저장
morphed_sentences = [okt.nouns(sentence.strip()) for sentence in sentences if sentence.strip()]

# 결과 파일 저장
with open('nouns.txt', 'w') as output_file:
    for i, nouns in enumerate(morphed_sentences, 1):
        output_file.write(f"{str(nouns)}\n")

print("저장 완료")
```

```python
output_folder = "text_split"
os.makedirs(output_folder, exist_ok=True)  # 폴더가 없으면 생성

with open('data.txt', 'r', encoding='utf-8') as file:
    for i in range(34):
        content = file.read(10000)  # 매번 10,000자씩 읽기
        if not content:  # 더 이상 읽을 내용이 없으면 종료
            break

        # 파일을 text_split 폴더 안에 저장
        filename = os.path.join(output_folder, f"{i+1}.txt")
        with open(filename, 'w', encoding='utf-8') as new_file:
            new_file.write(content)

        print(f"Created: {filename}")  # 생성된 파일 경로 출력
#336020
# 만자 정도는 인식을 한다.
# -> 파일을 34개로 나누자
# 일단 나누기는 완료
```

```python
# 일단 문장수는 했으니 
# 요소들을 추출 해야 한다.

output_text_processfolder = "text_process"
os.makedirs(output_text_processfolder, exist_ok=True)  # 폴더가 없으면 생성

for i in range(34):
    file = open(f'text_split/{i+1}.txt','r')
    content = file.read()
    sentence = okt.morphs(content)
    print(f"{i+1}번쨰 읽기는 함")
    with open(f"text_process/{i+1}.pkl", "wb") as file:
        pickle.dump(sentence, file)
        print(f"{i+1}번쨰 넣음")
```

```python
# 이제는 뭐를 해야 할까?
# csv 파일로 저장을 하자
output_csv = "word_counting.csv"

word_counts = Counter()

for sen in range(34):
    with open(f"text_process/{sen+1}.pkl", "rb") as file:
        
        loaded_data = pickle.load(file)
    print(f"{sen+1}번쨰 완료")
    content = loaded_data
    word_counts.update(content)
# 단어 빈도수를 DataFrame으로 변환
df = pd.DataFrame(word_counts.items(), columns=["단어", "빈도수"])

# 빈도수 기준으로 정렬 (내림차순)
df = df.sort_values(by="빈도수", ascending=False)

# CSV 파일로 저장
df.to_csv(output_csv, index=False)

print(f"✅ 단어 빈도수 분석 완료! CSV 파일 저장: {output_csv}")
```
